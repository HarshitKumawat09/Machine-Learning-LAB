```python
import pandas as pd

# Read the CSV dataset
# Treat '?' as missing values
df = pd.read_csv("crx.data", header=None, na_values='?')

# Dataset size before cleaning
rows_before, cols_before = df.shape
print("Dataset size before cleaning:")
print(f"Rows: {rows_before}, Columns: {cols_before}")

# Remove rows containing missing values
df_cleaned = df.dropna()

# Dataset size after cleaning
rows_after, cols_after = df_cleaned.shape
print("\nDataset size after cleaning:")
print(f"Rows: {rows_after}, Columns: {cols_after}")

# Number of rows removed
print("\nNumber of rows removed due to missing values:")
print(rows_before - rows_after)
```

```python
import pandas as pd

# Load the dataset
# Replace 'data.csv' with your actual file path
df = pd.read_csv('crx.data')

print("Original Dataset:")
print(df.head())

# Identify numerical and categorical columns
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns
categorical_cols = df.select_dtypes(include=['object']).columns

# Replace missing numerical values with mean (or median)
for col in numerical_cols:
    mean_value = df[col].mean()
    df[col].fillna(mean_value, inplace=True)

# Replace missing categorical values with a constant
for col in categorical_cols:
    df[col].fillna("Unknown", inplace=True)

print("\nDataset after handling missing values:")
print(df.head())

# Verify that no missing values remain
print("\nMissing values count after cleaning:")
print(df.isnull().sum())
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("crx.data", header=None, na_values='?')

print("Original Dataset Preview:")
print(df.head())

# Example of inconsistent values:
# Gender column may contain 'm', 'M', 'f', 'F'
# Approval column may contain '+', '-', 'yes', 'no'

# Replace inconsistent values using replace()
df.replace({
    'm': 'Male',
    'M': 'Male',
    'f': 'Female',
    'F': 'Female',
    '+': 'Approved',
    '-': 'Rejected',
    'yes': 'Approved',
    'no': 'Rejected'
}, inplace=True)

print("\nDataset After Replacing Inconsistent Values:")
print(df.head())

# Verify corrected values using value counts
print("\nVerification (Value Counts):")
print("Gender Column:")
print(df[1].value_counts(dropna=False))

print("\nApproval Status Column:")
print(df[15].value_counts(dropna=False))
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("AirQualityUCI_cleaned.csv")

# Combine Date and Time into DateTime
df['DateTime'] = pd.to_datetime(
    df['Date'] + ' ' + df['Time'],
    errors='coerce'
)

# -------- Line Plot (Trend over Time) --------
plt.figure()
plt.plot(df['DateTime'], df['CO(GT)'])
plt.xlabel("Date & Time")
plt.ylabel("CO Level")
plt.title("CO Level Trend Over Time")
plt.show()

# -------- Bar Plot (Categorical Comparison) --------
# Compare average CO level by month
df['Month'] = df['DateTime'].dt.month
monthly_avg = df.groupby('Month')['CO(GT)'].mean()

plt.figure()
plt.bar(monthly_avg.index, monthly_avg.values)
plt.xlabel("Month")
plt.ylabel("Average CO Level")
plt.title("Average CO Level by Month")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Load dataset
df = pd.read_csv("AirQualityUCI_cleaned.csv")

# Combine Date and Time into DateTime
df['DateTime'] = pd.to_datetime(
    df['Date'] + ' ' + df['Time'],
    errors='coerce'
)

# -------- Line Plot (Trend over Time) --------
plt.figure(figsize=(12, 6))
plt.plot(df['DateTime'], df['CO(GT)'])
plt.xlabel("Date & Time")
plt.ylabel("CO Level")
plt.title("CO Level Trend Over Time")

# Fix overlapping dates
plt.xticks(rotation=45)
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.AutoDateLocator())
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

plt.tight_layout()
plt.show()

# -------- Bar Plot (Categorical Comparison) --------
# Compare average CO level by month
df['Month'] = df['DateTime'].dt.month
monthly_avg = df.groupby('Month')['CO(GT)'].mean()

plt.figure(figsize=(8, 5))
plt.bar(monthly_avg.index, monthly_avg.values)
plt.xlabel("Month")
plt.ylabel("Average CO Level")
plt.title("Average CO Level by Month")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the cleaned dataset
df = pd.read_csv("AirQualityUCI_cleaned.csv")

# -------- Boxplot --------
plt.figure(figsize=(8, 5))
box = plt.boxplot(
    [df['CO(GT)'], df['NO2(GT)']],
    labels=['CO(GT)', 'NO2(GT)'],
    patch_artist=True
)

# Add colors
colors = ['lightgreen', 'lightblue']
for patch, color in zip(box['boxes'], colors):
    patch.set_facecolor(color)

plt.xlabel("Pollutants")
plt.ylabel("Concentration Level")
plt.title("Boxplot of Air Pollutants")
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```
