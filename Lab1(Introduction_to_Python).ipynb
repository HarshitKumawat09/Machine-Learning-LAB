```python
import numpy as np

# Create a NumPy array using a Python list
my_list = [10, 20, 30, 40, 50]
array = np.array(my_list)

# Display the array
print("NumPy Array:", array)

# Print the shape of the array
print("Shape of the array:", array.shape)

# Print the data type of the array
print("Data type of the array:", array.dtype)
```

```python
import numpy as np

# Create a 3×4 matrix of zeros
zeros_matrix = np.zeros((3, 4))

# Create a 2×5 matrix of ones
ones_matrix = np.ones((2, 5))

# Print the matrices
print("3×4 Matrix of Zeros:")
print(zeros_matrix)
print("Shape:", zeros_matrix.shape)

print("\n2×5 Matrix of Ones:")
print(ones_matrix)
print("Shape:", ones_matrix.shape)
```

```python
import numpy as np

# Create a NumPy array with numbers from 1 to 12
original_array = np.arange(1, 13)

# Reshape the array into a 3×4 matrix
reshaped_array = original_array.reshape(3, 4)

# Display the arrays and their shapes
print("Original Array:")
print(original_array)
print("Shape:", original_array.shape)

print("\nReshaped 3×4 Matrix:")
print(reshaped_array)
print("Shape:", reshaped_array.shape)
```

```python
import numpy as np

# Create a NumPy array of 10 random integers between 1 and 100
arr = np.random.randint(1, 101, 10)

# Display the array
print("Random Array:")
print(arr)

# Find minimum, maximum, mean, and sum
print("Minimum value:", arr.min())
print("Maximum value:", arr.max())
print("Mean value:", arr.mean())
print("Sum of values:", arr.sum())
```

```python
import pandas as pd

# Create a dictionary with student names and marks
data = {
    'Student_Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Marks': [85, 90, 78, 88]
}

# Create a Pandas DataFrame
df = pd.DataFrame(data)

# Display the DataFrame
print("Student DataFrame:")
print(df)

# Print data types of each column
print("\nData Types of Each Column:")
print(df.dtypes)

```
